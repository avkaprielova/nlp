```python
from nltk.parse import DependencyGraph
```

```python
def trees(file):
    trees = []
    
    with open(file, 'r', encoding = 'utf-8') as f:
        parsed_sents = f.read().split('\n\n')
        
        for sent in parsed_sents:
            tree = [line for line in sent.split('\n') if line and line[0] != '#']
            trees.append('\n'.join(tree))
    return trees
```

```python
# заведем функцию, которая будет собирать (максимальное) поддерево, корнем которого является заданная вершина
_FILTER_RELS = ['punct', 'conj', 'parataxis']
def get_subtree(nodes, node):
    if not nodes[node]['deps']:
        return [node]
    else:
        return [node] + [get_subtree(nodes, dep) for rel in nodes[node]['deps'] 
                         if rel not in _FILTER_RELS
                         for dep in nodes[node]['deps'][rel]]
```

```python
# чтобы потом получить просто индексы для выделения групп, нужно разобраться со вложенностью
def flatten(l):
    flat = []
    for el in l:
        if not isinstance(el, list):
            flat.append(el)
        else:
            flat += flatten(el)
    return flat
```

```python
def results(trees):
    result = []
    for t in trees:
        g = DependencyGraph(t, top_relation_label='root')
        np_list = []
        for n in g.nodes:
            if g.nodes[n]['ctag'] == 'NOUN':
                np = list(sorted(flatten(get_subtree(g.nodes, n))))
                np_list.append(
                    ' '.join([g.nodes[i]['word'] for i in np])
                )
        result.append(np_list)
    return result
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def get_tfidf_keywords(tfidf_vectorizer, text, top_n = 10):
    feature_names = vectorizer.get_feature_names()
    tfidf_vector = vectorizer.transform([text])
    feature_index = tfidf_vector.nonzero()[1]
    tfidf_scores = zip(feature_index, [tfidf_vector[0, x] for x in feature_index])
    ws = [(feature_names[i], s) for (i, s) in tfidf_scores]  # /sum_counts[0, cnt_features.index(feature_names[i])]
    ws.sort(key=lambda x: x[1], reverse = True)
    return(ws[:top_n])
```

**Оценить точность, полноту, F-меру выбранных методов относительно имеющейся разметки**

```python
import numpy as np
```

```python
def calculate_metrics(keywords, gold_standard):
    gold_standard = [x.lower() for x in gold_standard]
    recall = len(set([x[0] for x in keywords]) & set(gold_standard))/len(gold_standard) # полнота
    precision = len(set([x[0] for x in keywords]) & set(gold_standard))/len(keywords) # точность
    if precision + recall != 0:
        fscore = 2*precision*recall/(precision + recall)
    else:
        fscore = 0
    print('Recall:', recall)
    print('Precision:', precision)
    print('F-score:', fscore)
    return recall, precision, fscore
```
